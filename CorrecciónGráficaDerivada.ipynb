```python
# Graficar derivada de la función del movimiento armónico
# Autor: Paz Córdova Emmanuel
#Fecha: Septiembre 2, 2021

import matplotlib.pyplot as plt
import numpy as np 
import math

def f1(A, T, t):
    return A*np.sin((2*np.pi/T)*t)

def f2(A, T, texa):
    return A*(2*np.pi/T)*np.cos((2*np.pi/T)*texa)      

t1=np.arange(0, 10.00, 0.1)                  

A=float(input("Ingrese la amplitud del oscilador en metros: "))
T=float(input("Ingrese el periodo del oscilador en segundos: "))
#t=float(input("Ingrese el punto del tiempo en segundos donde quiere calcular la derivada: "))
h=float(input("Ingrese el valor del incremento: "))

plt.plot(t1, f1(A, T, t1))
plt.plot(t1, (f1(A, T, t1+h)-f1(A, T, t1))/h)
plt.plot(t1, f2(A, T, t1))
plt.show()
```
